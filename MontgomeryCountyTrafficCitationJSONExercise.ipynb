```python
datels=[] #Date of Stop
timels=[] #Time of Stop
agencyls=[] #Agency
subagencyls=[] #Subagency
descripls=[] #Description
locls=[] #Location
latls=[] #Latitude
longls=[] #Longitude
vehtypels=[] #Vehicle Type
vehyearls=[] #Vehicle Year
vehmakels=[] #Vehicle Make
vehmodells=[] #Vehicle Model
vehcolorls=[] #Vehicle Color
violtypels=[] #Violation Type
racels=[] #Race
genderls=[] #Gender
drivstls=[] #Driver's State
drivlicstls=[] #Driver's License State
arresttypels=[] #Arrest Type
```

```python
citations = data['data']

for citation in citations: #Need to verify correct terms to fill the append()s
    datels.append(citation[8])
    timels.append(citation[9])
    agencyls.append(citation[10])
    subagencyls.append(citation[11])
    descripls.append(citation[12])
    locls.append(citation[13])
    latls.append(citation[14])
    longls.append(citation[15])
    vehtypels.append(citation[27])
    vehyearls.append(citation[28])
    vehmakels.append(citation[29])
    vehmodells.append(citation[30])
    vehcolorls.append(citation[31])
    violtypels.append(citation[32])
    racels.append(citation[36])
    genderls.append(citation[37])
    drivstls.append(citation[39])
    drivlicstls.append(citation[40])
    arresttypels.append(citation[41])
    
datels[:5]
```
